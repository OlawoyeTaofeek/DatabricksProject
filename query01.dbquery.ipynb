```sql
SELECT * FROM customer;

-- Check for null
SELECT *
FROM customer
WHERE first_name IS NULL or last_name is NULL;

-- Check for space around first_name, last_name, email
SELECT *
FROM customer 
WHERE TRIM(first_name) != first_name 
     or TRIM(last_name) != last_name 
     or TRIM(email) != email;

-- Count Customers by Gender
SELECT gender, 
       COUNT(*) AS number_of_customers
FROM customer
GROUP BY gender
order by number_of_customers desc;

SELECT DISTINCT(loyalty_tier)
from customer;

-- Find the Number of Customers in Each Loyalty Tier
SELECT loyalty_tier,
      COUNT(1)
FROM customer
GROUP BY loyalty_tier
ORDER BY COUNT(1) DESC;

-- List Customers Who Signed Up in 2024
SELECT *
FROM customer
WHERE signup_date BETWEEN '2024-01-01' AND '2024-12-31';

SELECT *
FROM customer
WHERE YEAR(signup_date) = 2024;


-- Find the Most Common First Names
/*
  Calculate Customer Distribution by Loyalty Tier and Gender

  Goal: Multi-column grouping.
*/
SELECT loyalty_tier,
        gender,
        COUNT(*) as number_of_customers
from customer
group by loyalty_tier, gender
order by number_of_customers desc;

-- Take a good look at the sales table
SELECT *
FROM sales;

-- Return Table
SELECT *
FROM returns;




```
